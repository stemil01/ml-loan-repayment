```python
import os

import numpy as np
import pandas as pd

from sklearn import linear_model
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
dataset_dir = "dataset"
```

```python
data_df = pd.read_csv(os.path.join(dataset_dir, "merged_data.csv"))
```

```python
X = data_df.drop(columns=["TARGET"])
y = data_df["TARGET"]
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(X_train_val)
X_train_val_s = scaler.transform(X_train_val)
X_test_s = scaler.transform(X_test)
```
